```python
import pandas as pd
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import accuracy_score
from scipy.stats import entropy
import time
import matplotlib.pyplot as plt

# 讀取資料
data = pd.read_csv('banana.csv')

# 提取特徵資料和標籤
X = data.drop(columns=['class'])
y = data['class']
```

# K-means

```python
# K-means 分群
start_time = time.time()
kmeans = KMeans(n_clusters=2)
kmeans_pred = kmeans.fit_predict(X)
kmeans_time = time.time() - start_time

# 計算時間
print("K-means Time:", kmeans_time)

# 計算 SSE
kmeans_sse = kmeans.inertia_
print("K-means SSE:", kmeans_sse)

# 計算 Accuracy
kmeans_accuracy = accuracy_score(y, kmeans_pred)
print("K-means Accuracy:", kmeans_accuracy)

# 計算 Entropy
label_probs = pd.Series(kmeans_pred).value_counts(normalize=True)
kmeans_entropy = entropy(label_probs)
print("K-means Entropy:", kmeans_entropy)

# 繪製 K-means 分群圖表
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=kmeans_pred, cmap='viridis', marker='o')
plt.title('K-means Clustering')
plt.show()
```

# AgglomerativeClustering

```python
# 階層式分群
start_time = time.time()
agg = AgglomerativeClustering(n_clusters=2)
agg_pred = agg.fit_predict(X)
agg_time = time.time() - start_time

# 計算時間
print("Agglomerative Clustering Time:", agg_time)

# 計算 SSE（階層式分群無法直接計算SSE）

# 計算 Accuracy
agg_accuracy = accuracy_score(y, agg_pred)
print("Agglomerative Clustering Accuracy:", agg_accuracy)

# 計算 Entropy
label_probs = pd.Series(agg_pred).value_counts(normalize=True)
agg_entropy = entropy(label_probs)
print("Agglomerative Clustering Entropy:", agg_entropy)

# 繪製 階層式分群 圖表
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=agg_pred, cmap='viridis', marker='o')
plt.title('Agglomerative Clustering')
plt.show()
```

# DBSCAN

```python
# DBSCAN 分群
start_time = time.time()
dbscan = DBSCAN(eps=0.04, min_samples=2)
dbscan_pred = dbscan.fit_predict(X)
dbscan_time = time.time() - start_time

# 計算時間
print("DBSCAN Time:", dbscan_time)

# 計算 SSE（DBSCAN無法直接計算SSE）

# 計算 Accuracy
dbscan_accuracy = accuracy_score(y, dbscan_pred)
print("DBSCAN Accuracy:", dbscan_accuracy)

# 計算 Entropy
label_probs = pd.Series(dbscan_pred).value_counts(normalize=True)
dbscan_entropy = entropy(label_probs)
print("DBSCAN Entropy:", dbscan_entropy)

# 繪製 DBSCAN 分群圖表
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=dbscan_pred, cmap='viridis', marker='o')
plt.title('DBSCAN Clustering')
plt.show()
```

```python
def plot_clusters(X, pred_labels, title, markers):
    plt.figure(figsize=(8, 6))

    for label in set(pred_labels):
        cluster_points = X[pred_labels == label]
        plt.scatter(cluster_points.iloc[:, 0], cluster_points.iloc[:, 1],
                    marker=markers[label % len(markers)], edgecolors='k', s=50, label=f'Cluster {label}')

    plt.title(title)
    plt.legend()
    plt.show()


# 印出結果
markers = ['+', 'o']

# 使用更新後的plot_clusters函數繪製分群結果
plot_clusters(X, kmeans_pred, "K-means Clustering", markers)
plot_clusters(X, agg_pred, "Agglomerative Clustering", markers)
plot_clusters(X, dbscan_pred, "DBSCAN Clustering", markers)
```

```python
import itertools
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.metrics import accuracy_score
from scipy.stats import entropy
import time

# 初始化最佳參數和指標
best_accuracy = 0.0
best_entropy = float('inf')
best_params = None

# 定義 eps 和 min_samples 的範圍
eps_range = list(np.arange(0.01, 0.06, 0.01))  # 從 0.01 到 0.05，間隔 0.01
min_samples_range = list(range(1, 51))  # 從 1 到 50 的範圍

# 產生所有可能的參數組合
parameter_combinations = list(itertools.product(eps_range, min_samples_range))

# 使用迴圈進行遍歷
for eps, min_samples in parameter_combinations:
    # 在這裡放入你的程式碼，使用 eps 和 min_samples 這兩個參數
    start_time = time.time()
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    dbscan_pred = dbscan.fit_predict(X)
    dbscan_time = time.time() - start_time

    # 計算 Accuracy
    dbscan_accuracy = accuracy_score(y, dbscan_pred)

    # 計算 Entropy
    label_probs = pd.Series(dbscan_pred).value_counts(normalize=True)
    dbscan_entropy = entropy(label_probs)

    # 如果找到更好的 Accuracy 或更低的 Entropy，更新最佳參數和指標
    if dbscan_accuracy > best_accuracy or (dbscan_accuracy == best_accuracy and dbscan_entropy < best_entropy):
        best_accuracy = dbscan_accuracy
        best_entropy = dbscan_entropy
        best_params = {'eps': eps, 'min_samples': min_samples}

# 印出最佳參數和對應的 Accuracy、Entropy
print("Best Parameters:", best_params)
print("Best Accuracy:", best_accuracy)
print("Best Entropy:", best_entropy)
```
